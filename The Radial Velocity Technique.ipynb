# The Radial Velocity Technique

```python
import math
import matplotlib.pyplot as plt
```

```python
rad_v= []
theta_v= []
for i in range(0,720):
 theta_temp=i
 rad_temp = ((2*3.14*0.05*(math.sin(math.radians(60))))/(5*(math.sqrt(1-(0)))))*((math.cos(math.radians(i)))+(0*(math.cos(math.radians(0)))))
 rad_v.append(rad_temp)
 theta_v.append(theta_temp)
```

```python
plt.plot(theta_v, rad_v)
```
