```python
import math
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# Calculation of a using Orbital Time Period P using Kepler's Third Law
M = 1.988*(math.pow(10,30)) #Stellar Mass equal to Solar Mass in Kilograms 
G = 6.674*(math.pow(10,-11)) #G is expressed in Seconds, Kilogram
e = 0.9 #Eccentricity
inclination = 30
w = 90 #phase
```

```python
rad_v= []
theta_v= []
dt = []
i_m= []
dtt = 0;
a = 1
for i in range(1,720):
 rad_temp = ((2*3.14*a*(math.sin(math.radians(inclination))))/(5*(math.sqrt(1-(e*e)))))*((math.cos(math.radians(i+w)))+(e*(math.cos(math.radians(w)))))
 rad_v.append(rad_temp) #list with radial velocities
 r_temp = ((a*(1-(e*e)))/(1+(e*(math.cos(math.radians(i))))))
 c = G*M*a*(1-(e*e))
 w_temp= ((math.pow(c,1/2))/(r_temp*r_temp))
 dt_temp = ((i*3.14159265)/180)/w_temp #dt = dtheta/angular velocity
 dt.append((dt_temp))

time_temp = 0;
time = []
for i in range(1,720):
 for j in range(1,i):
    time_temp = time_temp+dt[j]
    
 time.append(time_temp)
 time_temp = 0;
```

```python
plt.plot(time,rad_v)
plt.xlabel("dt")
plt.ylabel("Radial Velocity")
plt.show
```
