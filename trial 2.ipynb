```python
import math
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
dtheta = []
i = 0
while i <= 720:
    dtheta.append(i)
    i += 0.00027777777
```

```python
# Calculation of a using Orbital Time Period P using Kepler's Third Law
M = 1.988*(math.pow(10,30)) #Stellar Mass equal to Solar Mass in Kilograms 
G = 6.674*(math.pow(10,-11)) #G is expressed in Seconds, Kilogram
```

```python
e = 0 #Eccentricity
p_t = (60*60*24*365) #p_t: One Earth Year expressed in seconds 
p = 5*p_t # multiplying Earth year in seconds with the orbital Period 
inclination = 60
a_temp = ((p*p*G*M)/(4*3.14*3.14)) #calcuating a cube using Keplers Third Law
a = math.pow(a_temp,1/3) #calculating a
w = 90 #phase
```

```python
rad_v= []
theta_v= []
dt = [0.0]
for i in dtheta:
 rad_temp = ((2*3.14*0.05*(math.sin(math.radians(inclination))))/(p*(math.sqrt(1-(e*e)))))*((math.cos(math.radians(i+w)))+(e*(math.cos(math.radians(w)))))
 rad_v.append(rad_temp) #list with radial velocities
 r_temp = ((a*(1-(e*e)))/(1+(e*(math.cos(math.radians(i))))))
 c = G*M*a*(1-(e*e))
 w_temp= ((math.pow(c,1/2))/(r_temp*r_temp))
 dt_temp = i/w_temp #dt = dtheta/angular velocity
 dt.append(dt_temp + dt[i])
```

```python
plt.plot(dt,rad_v)
plt.xlabel("dt")
plt.ylabel("Radial Velocity")  
plt.show
```
